```python
def likelihood(i_obs, i_model,f_obs,f_model):    chi_squared = sum((i_obs-i_model)**2+(f_obs-f_model))    return -chi_squareddef my_model(x1,y1,z1,x2,y2,z2,z1_ob,z2_ob):    m_x=(x2-x1)/(z2-z1)    m_y=(y2-y1)/(z2-z1)    b_x=x2-m_x*z1    b_y=y2-m_y*z2    sol1=(m_x*z1_ob+b_x)    sol2=(m_y*z2_ob+b_y)    solucion=[sol1,sol2]    return solucion
```
